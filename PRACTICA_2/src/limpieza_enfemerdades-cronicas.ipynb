```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/chronic-disease/U.S._Chronic_Disease_Indicators.csv')
```

```python
columnas = ['YearStart','LocationDesc','Topic']

df = df[columnas]
```

```python
df['Topic'].unique()
```

```python
# Crear un diccionario con las traducciones
translations = {
    'Alcohol': 'Alcohol',
    'Asthma': 'Asma',
    'Arthritis': 'Artritis',
    'Cancer': 'Cáncer',
    'Diabetes': 'Diabetes',
    'Overarching Conditions': 'Condiciones generales',
    'Tobacco': 'Tabaco',
    'Chronic Kidney Disease': 'Enfermedad renal crónica',
    'Chronic Obstructive Pulmonary Disease': 'Enfermedad pulmonar obstructiva crónica',
    'Cardiovascular Disease': 'Enfermedad cardiovascular',
    'Immunization': 'Inmunización',
    'Mental Health': 'Salud mental',
    'Nutrition, Physical Activity, and Weight Status': 'Nutrición, actividad física y estado de peso',
    'Reproductive Health': 'Salud reproductiva',
    'Oral Health': 'Salud oral',
    'Older Adults': 'Adultos mayores',
    'Disability': 'Discapacidad'
}

# Aplicar las traducciones al DataFrame
df['Topic'] = df['Topic'].map(translations)

# Verificar las traducciones
print(df['Topic'].unique())
```

```python
df.to_csv('enfermedades_cronicas_anos.csv',index=False, sep=';')
```
